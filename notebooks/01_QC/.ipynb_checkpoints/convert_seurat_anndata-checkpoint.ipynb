# Converting Seurat objects to h5ad objects

## Use case 1. Multiome object (Seurat -> h5ad)

```r
## NOTE
# 1) Make sure to open the Seurat file to check where the "raw" counts and data are saved. 
# 2) Check the number of genes, number of cells. Make sure that the converted object contains all genes, not scaled (usually Highly Variable Genes)

# Initialize the environment
# rm(list=ls())
library(Seurat) 
library(SeuratDisk)
```

```r
# Load the data (path to the R data file)
sobj <- readRDS("/data/yangjoon.kim/bruno/projects/data.science/yangjoon.kim/zebrahub_multiome/data/annotated_data/TDR118_processed.rds")
# sobj = load(input_dir_prefix)
sobj
```

```r
sobj@assays$RNA@counts
```

```r
sobj@meta.data
```

```r
# So, this method of using "Convert" function grabs  "scale.data" field preferentially from the Seurat object. 
# It's tricky to get rid of the scale.data field, thus we will copy into a new Seurat object without the scale.data field.
# FullSeurat = CreateSeuratObject(sobj@assays$RNA@counts, 
#             project = "CreateSeuratObject", assay = "RNA", 
#             names.field = 1, names.delim = "_", meta.data = sobj@meta.data)
```

```r
sobj@assays$RNA@data@x
```

```r
# Replace the original Seurat object's "scale.data" field with double [0x0] using the FullSeurat
sobj@assays$RNA@scale.data<-FullSeurat@assays$RNA@scale.data
# remove the FullSeurat object
# rm(FullSeurat)

# Make sure that we have the raw counts in the @x field
# Note that this should be checked manually for now, but we can think of adding a unit test for this (whether the sum is integer or not.)
sobj@assays$RNA@data@x <- sobj@assays$RNA@counts@x
# change the default assay to RNA here (if we want ATAC, or different assays, we can define it here.)
DefaultAssay(sobj) <- "RNA" 

# Set the directory to the output_dir where we will save the h5 and h5ad files
output_dir <- "/data/yangjoon.kim/bruno/projects/data.science/yangjoon.kim/zebrahub_multiome/data/annotated_data/"
setwd(output_dir)

# Save as seurat h5
# RNA object
SaveH5Seurat(sobj, filename = 'TDR118_processed_RNA.h5Seurat')

#convert to h5ad, writes to disk
Convert("TDR118_processed_RNA.h5Seurat", dest = "h5ad")
```

```r
# Set the directory to the output_dir where we will save the h5 and h5ad files
output_dir <- "/data/yangjoon.kim/bruno/projects/data.science/yangjoon.kim/zebrahub_multiome/data/annotated_data/"
setwd(output_dir)

#convert to h5ad, writes to disk
Convert("TDR118_processed.h5Seurat", dest = "h5ad")
```
