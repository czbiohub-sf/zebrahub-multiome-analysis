## subset the adata for NMP trajectories

```python
# import the csv file for the "alignedUMAP" coordinates for the NMP trajectory
umap_coords_nmps = pd.read_csv("/hpc/projects/data.science/yangjoon.kim/zebrahub_multiome/data/processed_data/01_Signac_processed/aligned_umap_coords_NMPs.csv", index_col=0)
umap_coords_nmps.head()
```

```python
# dictionary for data_id and timepoints
dict_datasets = {"0somites":"TDR126",
                 "5somites":"TDR127",
                 "10somites":"TDR128",
                 "15somites":"TDR118reseq",
                 "20somites":"TDR125reseq",
                 "30somites":"TDR124reseq"}
```

```python
# create a new directory to save the NMP subsetted datasets
os.makedirs("/hpc/projects/data.science/yangjoon.kim/zebrahub_multiome/data/processed_data/08_NMPs_subsetted/", exist_ok=True)
```

```python
umap_coords_sub["cell_type"].values
```

```python
# for each timepoint, load the adata, subset the object using the cell_id
adata_path = "/hpc/projects/data.science/yangjoon.kim/zebrahub_multiome/data/processed_data/01_Signac_processed/"

adata_sub_path = "/hpc/projects/data.science/yangjoon.kim/zebrahub_multiome/data/processed_data/08_NMPs_subsetted/"

for timepoint in dict_datasets.keys():
    print(timepoint)
    data_id = dict_datasets[timepoint]
    data_name = data_id.replace("reseq","")
    # load the adata
    adata = sc.read_h5ad(adata_path + f"{data_id}/{data_name}_processed_RNA.h5ad")
    print(adata)
    
    # subset the umap_coords
    umap_coords_sub = umap_coords_nmps[umap_coords_nmps.timepoint==timepoint]
    # reformat the cell_id
    umap_coords_sub.cell_id = umap_coords_sub.cell_id.str.split("_").str[0]
    
    # subset for the NMP trajectories
    adata_sub = adata[adata.obs_names.isin(umap_coords_sub.cell_id)]
    print(adata_sub)
    
    # add the aligned_UMAP coordinates into the adata_sub
    adata_sub.obsm["X_umap_aligned"] = umap_coords_sub[['UMAP_1', 'UMAP_2']].values
    # add the manual_annotation to the obs fields
    adata_sub.obs["manual_annotation"] = umap_coords_sub["cell_type"].values
    
    # Create a list of columns in adata.obs that contain the word 'prediction'
    cols_to_drop = [col for col in adata_sub.obs.columns if 'prediction' in col]

    # Drop these columns from adata.obs
    adata_sub.obs.drop(columns=cols_to_drop, inplace=True)
    
    # Replace "." with "_" between words (this is for Cellxgene compatibility)
    # Create a dictionary mapping old column names to new names where dots are replaced with underscores
    rename_dict = {col: col.replace('.', '_') for col in adata_sub.obs.columns}

    # Rename the columns in adata.obs using this dictionary
    adata_sub.obs.rename(columns=rename_dict, inplace=True)
    
    del adata_sub.raw
    
    sc.pl.embedding(adata_sub, basis="X_umap_aligned", color=["global_annotation","manual_annotation"])

    adata_sub.write_h5ad(adata_sub_path+f"{data_id}_NMPs.h5ad")
    
    
    
    
```

### generate alignedUMAP plots for each timepoint (with manual annotation)

```python
import seaborn as sns

# set the color palettes using "Set3"
# Cell types
cell_types = [
    'Epidermal', 'Lateral_Mesoderm', 'PSM', 'Neural_Posterior',
    'Neural_Anterior', 'Neural_Crest', 'Differentiating_Neurons',
    'Adaxial_Cells', 'Muscle', 'Somites', 'Endoderm', 'Notochord',
    'NMPs'
]

# Generate "Set3" color palette
set3_palette = sns.color_palette("Set3", n_colors=len(cell_types))

# Suppose the light yellow is the 10th color in the palette (9th index, as indexing is 0-based)
# and you want to replace it with teal color
teal_color = (0.0, 0.5019607843137255, 0.5019607843137255)  # RGB for teal
set3_palette[1] = teal_color  # Replace the light yellow with teal

# Assign colors to cell types
custom_palette = {cell_type: color for cell_type, color in zip(cell_types, set3_palette)}
# Change the color for 'NMPs' to a dark blue
custom_palette['NMPs'] = (0.12941176470588237, 0.4, 0.6745098039215687)  # Dark blue color
```
