## Pseudotime calculation using pySlingshot

- source: https://github.com/mossjacob/pyslingshot/blob/master/slingshot.ipynb 
- last updated: 05/14/2024




```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
from pyslingshot import Slingshot

import scanpy as sc
import anndata as ad
```

```python
adata = sc.read_h5ad("/hpc/projects/data.science/yangjoon.kim/zebrahub_multiome/data/processed_data/01_Signac_processed/integrated_RNA_ATAC_counts_RNA_formatted_excxg_daniocell_labels.h5ad")
adata
```

```python
manual_annotation = pd.read_csv("/hpc/projects/data.science/yangjoon.kim/zebrahub_multiome/data/processed_data/01_Signac_processed/annotations/annotation_ML.txt", sep="\t", index_col=0)
manual_annotation
```

```python
adata.obs["manual_annotation"] = manual_annotation
```

```python
adata
```

```python
manual_annotation.annotation_ML.unique()
```

```python
adata[adata.obs.manual_annotation=="unassigned"]
```

```python
sc.pl.embedding(adata, basis="X_wnn.umap", color="manual_annotation")
```

```python
sc.pl.embedding(adata, basis="X_wnn.umap", color="global_annotation")
```

```python
# Assuming 'manual_annotation' contains the categories as shown
categories = adata.obs['manual_annotation'].cat.categories

# Create a dictionary mapping each category to an integer
category_to_integer = {category: i for i, category in enumerate(categories)}

# Print the mapping to verify
print(category_to_integer)
```

```python
category_to_integer["tail_bud"]
```

```python
# Replace the categorical labels in the DataFrame with the mapped integers
adata.obs['manual_annotation_integer'] = adata.obs['manual_annotation'].map(category_to_integer)

# # Check the new column to ensure the mapping is correct
# print(adata.obs.head())
```

```python
# Convert categorical labels to integer codes
adata.obs['manual_annotation_integer'] = adata.obs['manual_annotation'].cat.codes

# Check the new column to ensure the mapping is correct and the datatype
print(adata.obs['manual_annotation_integer'].head())
print(adata.obs['manual_annotation_integer'].dtype)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
custom_xlim = (-12, 12)
custom_ylim = (-12, 12)
# plt.setp(axes, xlim=custom_xlim, ylim=custom_ylim)

slingshot = Slingshot(adata, celltype_key="manual_annotation_integer", 
                      obsm_key="X_wnn.umap", start_node=37, 
                      debug_level='verbose')

slingshot.fit(num_epochs=1, debug_axes=axes)
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(12, 4))
axes[0].set_title('Clusters')
axes[1].set_title('Pseudotime')
slingshot.plotter.curves(axes[0], slingshot.curves)
slingshot.plotter.clusters(axes[0], labels=np.arange(slingshot.num_clusters), s=4, alpha=0.5)
slingshot.plotter.clusters(axes[1], color_mode='pseudotime', s=5)
```

### get Pseudotime

```python
# NOTE: the Slingshot class has a property which has the pseudotime that is used to 
# color the plot above
pseudotime = slingshot.unified_pseudotime
pseudotime
```

```python
len(pseudotime)
```

### map the pseudotime into adata object

```python
adata
```

```python
adata.obs["Pseudotime"] = pseudotime
```

```python
adata.obs.head()
```

```python
sc.pl.embedding(adata, basis="X_umap_aligned",
                color="Pseudotime", vmax=20)
```

```python
# add the columns for lineages (mesoderm/neuroectoderm) - used later in CellOracle
Lineage_Meso = ["NMPs","PSM","Somites","Muscle"]
Lineage_NeuroEcto = ["NMPs", "Neural_Posterior", "Neural_Anterior"]

adata.obs["Lineage_Meso"] = adata.obs.manual_annotation.isin(Lineage_Meso)
adata.obs["Lineage_NeuroEcto"] = adata.obs.manual_annotation.isin(Lineage_NeuroEcto)
```

```python
adata.obs["Lineage_Meso"] = adata.obs.manual_annotation.isin(Lineage_Meso)
adata.obs["Lineage_NeuroEcto"] = adata.obs.manual_annotation.isin(Lineage_NeuroEcto)
```

```python
# Fill in Pseudotime for Lineage_Meso where Lineage_Meso is True
adata.obs['Pseudotime_Lineage_Meso'] = adata.obs['Pseudotime'].where(adata.obs['Lineage_Meso'], np.nan)

# Fill in Pseudotime for Lineage_NeuroEcto where Lineage_NeuroEcto is True
adata.obs['Pseudotime_Lineage_NeuroEcto'] = adata.obs['Pseudotime'].where(adata.obs['Lineage_NeuroEcto'], np.nan)
```

```python
adata.obs.head()
```

```python
adata.obs.to_csv("/hpc/projects/data.science/yangjoon.kim/zebrahub_multiome/data/processed_data/08_NMPs_subsetted/TDR118reseq_NMPs_slingshot.csv")
```

## testing pySlingshot with different embeddings (UMAPs with different computations)

```python
adata_20somites = sc.read_h5ad("/hpc/projects/data.science/yangjoon.kim/zebrahub_multiome/data/processed_data/08_NMPs_subsetted/TDR125reseq_NMPs.h5ad")
adata_20somites
```

```python
sc.pl.embedding(adata_20somites, basis="X_umap.atac", color="manual_annotation")
```

```python
sc.pl.embedding(adata_20somites, basis="X_umap.joint", color="manual_annotation")
```

```python
sc.pl.embedding(adata_20somites, basis="X_umap_aligned", color="manual_annotation")
```

```python
adata_20somites.layers["counts"] = adata_20somites.X.copy()
```

```python
# adata_20somites.X = adata_20somites.layers["counts"].copy()
```

```python
# pre-process the dataset and re-compute the UMAP
# log-normalization
sc.pp.normalize_total(adata_20somites, target_sum=1e4)
sc.pp.log1p(adata_20somites)

sc.pp.scale(adata_20somites, max_value=10)
sc.tl.pca(adata_20somites, svd_solver="arpack")
sc.pp.neighbors(adata_20somites, n_neighbors=10, n_pcs=40)
sc.tl.umap(adata_20somites)
```

```python
sc.pp.neighbors(adata_20somites, n_neighbors=10, n_pcs=40)
sc.tl.umap(adata_20somites, min_dist=0.05)
sc.pl.umap(adata_20somites, color="manual_annotation")
```

```python
# Assuming 'manual_annotation' contains the categories as shown
categories = adata_20somites.obs['manual_annotation'].cat.categories

# Create a dictionary mapping each category to an integer
category_to_integer = {category: i for i, category in enumerate(categories)}

# Print the mapping to verify
print(category_to_integer)
```

```python
# Replace the categorical labels in the DataFrame with the mapped integers
adata_20somites.obs['manual_annotation_integer'] = adata_20somites.obs['manual_annotation'].map(category_to_integer)

# # Check the new column to ensure the mapping is correct
# print(adata.obs.head())
```

```python
# Convert categorical labels to integer codes
adata_20somites.obs['manual_annotation_integer'] = adata_20somites.obs['manual_annotation'].cat.codes

# Check the new column to ensure the mapping is correct and the datatype
print(adata_20somites.obs['manual_annotation_integer'].head())
print(adata_20somites.obs['manual_annotation_integer'].dtype)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
custom_xlim = (-12, 12)
custom_ylim = (-12, 12)
# plt.setp(axes, xlim=custom_xlim, ylim=custom_ylim)

slingshot = Slingshot(adata_20somites, celltype_key="manual_annotation_integer", 
                      obsm_key="X_umap.atac", start_node=1, 
                      debug_level='verbose')

slingshot.fit(num_epochs=1, debug_axes=axes)
```
